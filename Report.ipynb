```python
import pandas as pd
```

```python
df = pd.read_csv('balance.txt', delim_whitespace=True)
```

```python
print (df['Ethnicity'].unique())
```

```python
print("The average income of Caucasians is", df[df.Ethnicity == "Caucasian"].loc[:,"Income"].mean())
print("The average income of Asians is", df[df.Ethnicity == "Asian"].loc[:,"Income"].mean())
print("The average income of African Americans is", df[df.Ethnicity == "African American"].loc[:,"Income"].mean())
```

```python
print("The average balance of Married individuals is", df[df.Married == "Yes"].loc[:,"Balance"].mean())
print("The average balance of Single individuals is", df[df.Married == "No"].loc[:,"Balance"].mean())
```

The average balance of Married individuals according to this dataset is lower than Single individuals.

```python
print ('The highest income in our dataset is:', df['Income'].max())
```

```python
print ('The lowest income in our dataset is:', df['Income'].min())
```

```python
print ('We have', df['Cards'].sum(), 'Cards in our dataset.')
```

```python
print ('We have information on', df[df.Gender == 'Female'].loc[:,"Balance"].count(), 'Females in our dataset.')
print ('We have information on', df[df.Gender == 'Male'].loc[:,"Balance"].count(), 'Males in our dataset.')
```
